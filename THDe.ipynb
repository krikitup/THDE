```python
delta = [1.7,1.8,1.9,2,2.1,2.2,2.3]
delta
```

```python
w_t = [0.1]*len(delta)
for x in range(len(delta)):
    w_t[x] = -1 + 2*(-2*delta[x] + 4)*t/(3*(1+t**2))
```

```python
for x in range(len(delta)):
    plt.plot(t,w_t[x])
    plt.legend(delta)
plt.xlabel('$t(Gyr)$')
plt.ylabel('$\omega_t$')
plt.savefig("wt")
```

```python
t_0 = 13.77
alpha = 2
H_0 = 73
C = 3
A = 1.1036
t_lapse = 705.8186241423017
```

```python
time = np.linspace(0,t_0,500)
A = 1.1036
H = 1/np.log(A*(1+time**2))
```

```python
rho_t = [0.01]*len(delta)
for i in range(len(delta)):
    rho_t[i] = C*H**(4-2*delta[i])
```

```python
for i in range(len(delta)):
    plt.plot(time, rho_t[i], )
plt.xlabel('$t(Gyr)$')
plt.ylabel('THDE density')
plt.legend(delta)
plt.savefig('rhot.jpg')
```

```python
rho_m = [0.1]*len(delta)
for i in range(len(delta)):
    rho_m[i] = 3*H**2 - rho_t[i]
    plt.plot(time, rho_m[i])
plt.xlabel('$t(Gyr)$')
plt.ylabel('Matter density')
plt.legend(delta)
plt.savefig('matter.jpg')
    
```

```python
p_t = [0.1]*len(delta)
for i in range(len(delta)):
    w_t[i] = -1 + alpha*time*(4 -2*delta[i])/(3*(1 + time**2))
    p_t[i] = rho_t[i]*w_t[i]
    plt.plot(time, p_t[i])
plt.xlabel('$t(Gyr)$')
plt.ylabel('THDE pressure')
plt.legend(delta)
plt.savefig('pressure.jpg')
```

```python
q = -1 + alpha*time/(1+ time**2)
r = q + q**2 - alpha*(1 -time**2)/(H*(1 + time**2)**2)
s = (r - 1)/(3*(q - 1/2))
```

```python
fig, ax = plt.subplots()
ax.plot( r, s, 'grey', label = '$r-s$')
ax.plot(r, [1]*len(r), 'g--', label = '$\Lambda- CDM$')
plt.xlabel('r')
plt.ylabel('s')
plt.legend()
plt.savefig("rszoom.jpg")
```
